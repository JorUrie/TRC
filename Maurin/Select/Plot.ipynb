```python
import os
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
# --------------------------------------------------------------------------------------------------
# Limpiando pantalla
os.system ("cls") 
# --------------------
# Llamando los archivos que contienen los datos
date = pd.read_csv('FechaCD-CM.csv', header = 0, usecols = [0])
date1 = pd.read_csv('FechaCD-CM.csv', header = 0, usecols = [1])#'Data-MS.csv',header = 0, usecols = ['Year'])
e1 = pd.read_csv('PhiCF-MS-SNAE.csv') #'AATB-MS.csv', header = 0, usecols = ['AATB'])
# Capturando datos
DateE = pd.DataFrame(date)
DateE1 = pd.DataFrame(date1)
E1E = pd.DataFrame(e1)
# Transformadolo a lista
DateL = DateE.values.tolist()
DateL1 = DateE1.values.tolist()
E1L = E1E.values.tolist()
#Generando valores zero para que sea un arreglo mas acomodado
DateA = []
DateB = []
E1A = np.zeros(len(date))
# Convirtiendo en arreglos
Date = np.array(DateL)
Date1 = np.array(DateL1)
E1 = np.array(E1L)
for i in range(len(Date)):
    DateA.extend([str(Date1[i]) + '-' + str(Date[i])])
    E1A[i] = E1[i]
# Dando formato a las fechas
chars = '[, ]'
DateB = []
for j in range(len(DateA)):
    DateB.append(DateA[j].translate(str.maketrans('', '', chars)))
# Seleccionando valores maximos de cada arreglo para usar valores relativos
E1max = np.amax(E1)
# Generando los vectores de valores relativos
E1R = (E1A/E1max)*100
plt.rcParams['axes.facecolor'] = 'white'
plt.plot(DateB, E1A, '.', linewidth = 1, color = 'darkred')
plt.xticks(np.arange(0, len(DateB), 100), rotation = 90, fontsize = 7)
plt.title('Modulation Factor SNAE Data with \n Force Field solution', fontsize = 18)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Counts/s', fontsize = 16)
plt.ylim(0, 2)
#plt.xlim(1957, 2020)
#plt.ylim(65, 110)
plt.grid()
plt.savefig('Modulation_Factor_SNAE_Data_with_Force_Field_solution_Maurin', bbox_inches = 'tight')
plt.show()
#plt.savefig('Sunspot.jpg')
```
