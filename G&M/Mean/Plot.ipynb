```python
import os
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
# --------------------------------------------------------------------------------------------------
# Limpiando pantalla
os.system ("cls") 
# --------------------
# Llamando los archivos que contienen los datos
date = pd.read_csv('FechaCFMS.csv')#,header = 0, usecols = ['Year'])
e1 = pd.read_csv('PhiCF-MS-P.csv')#,header = 0, usecols = ['SNAE'])

# Capturando datos
DateE = pd.DataFrame(date)
E1E = pd.DataFrame(e1)
# Transformadolo a lista
DateL = DateE.values.tolist()
E1L = E1E.values.tolist()
#Generando valores zero para que sea un arreglo mas acomodado
DateA = np.zeros(len(date))
E1A = np.zeros(len(date))
# Convirtiendo en arreglos
Date = np.array(DateL)
E1 = np.array(E1L)
for i in range(len(Date)):
    DateA[i] = Date[i]
    E1A[i] = E1[i]
# Seleccionando valores maximos de cada arreglo para usar valores relativos
E1max = np.amax(E1)
# Generando los vectores de valores relativos
E1R = (E1A/E1max)*100
plt.rcParams['axes.facecolor'] = 'FFFFCB' # 'E6E6FA'
plt.plot(DateA, E1A, '.', linewidth = 1)
plt.title('Factor de Modulación Campo de Fuerza. Promedio', fontsize = 18)
plt.xlabel('Fecha', fontsize = 16)
plt.ylabel('Cuentas/s', fontsize = 16)
plt.ylim(0, 2)
#plt.xlim(1957, 2020)
plt.grid()
plt.show()
```
